# Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
query1 = "SELECT *, CASE WHEN ((WEEKDAY(rental_Date) >= 1) AND (WEEKDAY(rental_date) <= 5)) then 'workday' WHEN (WEEKDAY (rental_Date) = 6 OR WEEKDAY(rental_Date) = 7) then 'weekend'\
ELSE 'no_info' END AS day_type FROM rental"

#query1 = "SELECT *, CASE WHEN EXTRACT(WEEKDAY FROM rental_date) IN (0, 6) THEN 'weekend' ELSE 'workday' END AS day_type FROM rental"

data1 = pd.read_sql_query(query1, engine)
display(data1.head())
```

# Get all films with ARMAGEDDON in the title.

```python
data1 = pd.read_sql_query("SELECT * FROM film WHERE title LIKE '%%ARMAGEDDON%%'", engine)
data1
```

# Get 10 the longest films.

```python
data1 = pd.read_sql_query("SELECT * FROM film ORDER BY length DESC LIMIT 10", engine)
data1
```

# How many films include Behind the Scenes content?

```python
data1 = pd.read_sql_query("SELECT count(*) AS 'Films included Behind The Scene Content' FROM film WHERE special_features LIKE '%%Behind the Scenes%%'", engine)
data1
```

# Which kind of movies (rating) have a mean duration of more than two hours?

```python
data1 = pd.read_sql_query("SELECT rating, round(avg(length),2) AS mean_duration FROM film GROUP BY rating HAVING mean_duration>120", engine)
data1
```

# Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
data1 = pd.read_sql_query("SELECT Title, Length, RANK() OVER (ORDER BY length) AS 'rank_duration' FROM film WHERE length != 0 AND length IS NOT NULL", engine)
data1
```
